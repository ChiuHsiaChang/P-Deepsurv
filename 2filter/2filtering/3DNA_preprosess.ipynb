```python
clinical_df=pd.read_csv('D:\\2020\\資料科學\\final-project\\clinical\\clinical_filter2.csv',engine='python')
```

```python
clinical_df=clinical_df.rename(columns={'case_submitter_id':'id','vital_status':'fstatus'})
```

```python
clinical_df.head()
```

```python
cate=clinical_df.columns[clinical_df.dtypes=="object"].tolist()
cate
```

```python
cate.remove('id')
cate.remove('fstatus')
```

```python
oe = OrdinalEncoder()
clinical_df.loc[:,cate] = oe.fit_transform(clinical_df.loc[:,cate])
clinical_df.head()
```

```python
clinical_df['fstatus'] = clinical_df['fstatus'].map({'Alive': 1, 'Dead': 0})
clinical_df.head()
```

```python
j=clinical_df.shape[0]
for i in range(j):
    if (pd.isnull(clinical_df.ix[i,'age_at_diagnosis'])==True):
        clinical_df.ix[i,'age_at_diagnosis']=clinical_df.ix[i,'age_at_index']*365
```

```python
# format(float(futime)/365,'.4f')
clinical_df['futime']=clinical_df['futime'].apply(lambda x: x/365)
clinical_df['age_at_diagnosis']=clinical_df['age_at_diagnosis'].apply(lambda x: x/365)
clinical_df.head()
```

```python
clinical_df.isnull().sum()
```

```python
clinical_df=clinical_df.drop(columns=['age_at_index'])
```

```python
clinical_df.to_csv('D:\\2020\\資料科學\\final-project\\clinical\\clinical_filter_final.csv',index=False)
```
